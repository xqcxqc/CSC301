```python
# import libraries

import numpy as np
import random
import math 
```

```python
# method 1: numpy.cos(x)

# the derivative functin of sin(x) is cos(x), thus we use np.cos(x)
x = 1.0
exact = np.cos(x)
print (f"The result computed by np.cos(x) is: {exact}")
```

```python
# method 2: 3-point endpoint formula:

def Three_PointEndpoint(x,h):
    derivative = ((-3*(np.sin(x)))+(4*(np.sin(x+h)))-(np.sin(x+(2*h))))/((2*h))
    return derivative

h=0.1
T_end_ary = []

# vary h to find the value of h that leads to the most accurate results
for i in range(100):
    T_end = Three_PointEndpoint(x,h)
    T_end_ary.append(T_end)
    
    print (i,f". Result from three end point is: {T_end}.")
    print (f"   When h is ",h)
    h = h/10

```

```python
# Determine the absolute and relative errors for each of your calculations
# we assume that using np.cos(x) as the exact result.

#abs  : current - exact
T_abs = []
#rel  : abs/exact
T_rel = []

h = 0.1
print("For three-point endpoint formula, the errors are shown as following when h is different: \n")

for i in range(len(T_end_ary)):
    a = abs(T_end_ary[i] - exact)
    r = a/exact
    T_abs.append(a)
    T_rel.append(r)
    print("\n","When h is " , h)
    print("Absolute error: ", a)
    print("Relative error: ", r)
    h=h/10
```

```python
smallest_err_T = min(T_abs)
exponent_T = -(T_abs.index(smallest_err_T))
th = 0.1 * math.pow(10,exponent_T)

smallest_err_T2 = min(T_rel)
exponent_T2 = -(T_rel.index(smallest_err_T2))
th2 = 0.1 * math.pow(10,exponent_T2)

print( "The smallest abslute error is ", smallest_err_T)
print( "When h is ", th)

print( "The smallest relative error is ", smallest_err_T2)
print( "When h is ", th2)

print("Thus, the most accurate result for this method is ", T_end_ary[-exponent_T])
```
